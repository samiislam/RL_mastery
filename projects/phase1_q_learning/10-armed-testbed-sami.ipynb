```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(37)

# This was a set of 2000 randomly generated k-armed bandit problems with k = 10. For each bandit problem,
num_sets = 2000
num_steps = 1000
num_arms = 10
epsilon = 0.0
expected_reward = {0.0:0.0, 0.1:0.0, 0.01:0.0}

for epsilon in [0.0, 0.1, 0.01]:
    for set in range(num_sets):
        # 1 set = 1 bandit problem
        q_star_of_a = np.random.normal(0, 1, num_arms)
        # Sum of rewards for an action
        sum_reward_for_action = [0.0 for r in range(10)]
        num_action_taken = [0 for r in range(10)]
        Q_t = [0.0 for r in range(10)]

        for t in range(num_steps):
                    
            # 1000 time steps = experience of 1 bandit problem

            # Select Action A_t at time t
            if np.random.random() < epsilon:
                action_t = np.random.randint(0, num_arms) # explore
            else:
                action_t = np.argmax(Q_t).item() # exploit
            
            sum_of_rewards_when_a_taken_prior_to_t = sum_reward_for_action[action_t]
            num_of_times_a_taken_prior_to_t = num_action_taken[action_t]
            
            if num_of_times_a_taken_prior_to_t == 0:
                Q_t[action_t] = 0.0
            else:
                Q_t[action_t] = sum_of_rewards_when_a_taken_prior_to_t / num_of_times_a_taken_prior_to_t

            # Actual reward
            reward_t = np.random.normal(q_star_of_a[action_t], 1)
            sum_reward_for_action[action_t] += reward_t
            num_action_taken[action_t] += 1

        action_t = np.argmax(Q_t).item()
        expected_reward[epsilon] = (expected_reward[epsilon] + Q_t[action_t]) / (set + 1)
```
